##### 1. A set of one-dimensional data points is given to you: 5, 10, 15, 20, 25, 30, 35. Assume that k = 2 and that the first set of random centroid is 15, 32, and that the second set is 12, 30. ?
1. Using the k-means method, create two clusters for each set of centroid described above.
2. For each set of centroid values, calculate the SSE.

```python
from sklearn.cluster import KMeans
import numpy as np

# Given data points
data_points = np.array([5, 10, 15, 20, 25, 30, 35]).reshape(-1, 1)

# First set of centroids
centroids1 = np.array([[15], [32]])

# Second set of centroids
centroids2 = np.array([[12], [30]])

# Initialize KMeans with k=2 clusters
kmeans1 = KMeans(n_clusters=2, init=centroids1, n_init=1)
kmeans2 = KMeans(n_clusters=2, init=centroids2, n_init=1)

# Fit KMeans models
kmeans1.fit(data_points)
kmeans2.fit(data_points)

# Get cluster labels and centroids
labels1 = kmeans1.labels_
labels2 = kmeans2.labels_
centroids1 = kmeans1.cluster_centers_
centroids2 = kmeans2.cluster_centers_

# Calculate SSE for each set of centroids
sse1 = np.sum((data_points - centroids1[labels1])**2)
sse2 = np.sum((data_points - centroids2[labels2])**2)

print("SSE for first set of centroids:", sse1)
print("SSE for second set of centroids:", sse2)
```

This information helps businesses understand customer behavior, optimize product placement, and develop targeted marketing strategies.




##### 3. Give an example of the Apriori algorithm for learning association rules ?
**Ans:** Suppose we have a transactional dataset from a grocery store. The Apriori algorithm starts by identifying frequent itemsets (e.g., {milk}, {bread}, {milk, bread}) based on a minimum support threshold. Then, it generates association rules like {milk} -> {bread} by calculating confidence. If the confidence of {milk} -> {bread} exceeds a minimum threshold, it suggests that customers who buy milk also tend to buy bread. These rules help optimize product placement or create promotions (e.g., offering discounts on bread to customers who purchase milk) to increase sales and customer satisfaction.







##### 4. In hierarchical clustering, how is the distance between clusters measured? Explain how this metric is used to decide when to end the iteration ?
**Ans:** Distance between clusters is measured by various metrics (e.g., Euclidean, Manhattan). Iteration ends when clusters merge based on a predefined criterion (e.g., maximum linkage, average linkage) and distances exceed a threshold or form a predetermined number of clusters.





##### 5. In the k-means algorithm, how do you recompute the cluster centroids ?

**Ans:** In the k-means algorithm, cluster centroids are recomputed by taking the mean of all data points assigned to each cluster. This process involves calculating the average position of all data points in each cluster along each dimension. The resulting mean values represent the new centroids for the next iteration.

##### 7. Discuss the k-means algorithm's advantages and disadvantages ?

**Ans:**

**2.Disadvantages:**

##### 9. During your study, you discovered seven findings, which are listed in the data points below. Using the K-means algorithm, you want to build three clusters from these observations. The clusters C1, C2, and C3 have the following findings after the first iteration ?
- `C1: (2,2), (4,4), (6,6); C2: (2,2), (4,4), (6,6); C3: (2,2), (4,4),  `
- `C2: (0,4), (4,0), (0,4), (0,4), (0,4), (0,4), (0,4), (0,4), (0,  `
- `C3: (5,5) and (9,9) ` 

What would the cluster centroids be if you were to run a second iteration? What would this clustering's SSE be?

```python
import numpy as np

# Define data points for each cluster after the first iteration
C1 = np.array([[2, 2], [4, 4], [6, 6]])
C2 = np.array([[0, 4], [4, 0], [0, 4], [0, 4], [0, 4], [0, 4], [0, 4], [0, 4], [0, 4]])
C3 = np.array([[5, 5], [9, 9]])

# Compute centroids for each cluster
centroid_C1 = np.mean(C1, axis=0)
centroid_C2 = np.mean(C2, axis=0)
centroid_C3 = np.mean(C3, axis=0)

# Compute SSE for each cluster
sse_C1 = np.sum((C1 - centroid_C1) ** 2)
sse_C2 = np.sum((C2 - centroid_C2) ** 2)
sse_C3 = np.sum((C3 - centroid_C3) ** 2)

# Compute total SSE
total_sse = sse_C1 + sse_C2 + sse_C3

print("Centroid for C1 after second iteration:", centroid_C1)
print("Centroid for C2 after second iteration:", centroid_C2)
print("Centroid for C3 after second iteration:", centroid_C3)
print("SSE for C1 after second iteration:", sse_C1)
print("SSE for C2 after second iteration:", sse_C2)
print("SSE for C3 after second iteration:", sse_C3)
print("Total SSE after second iteration:", total_sse)
```

##### 10. In a software project, the team is attempting to determine if software flaws discovered during testing are identical. Based on the text analytics of the defect details, they decided to build 5 clusters of related defects. Any new defect formed after the 5 clusters of defects have been identified must be listed as one of the forms identified by clustering. A simple diagram can be used to explain this process. Assume you have 20 defect data points that are clustered into 5 clusters and you used the k-means algorithm ?

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic defect data points
X, _ = make_blobs(n_samples=20, centers=5, cluster_std=1.0, random_state=42)

# Perform k-means clustering
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Plot the defect data points
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', edgecolor='k', s=100)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', c='red', s=200, label='Centroids')
plt.title('Clustering of Software Defects')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```
